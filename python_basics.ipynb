Define a function validate_password that returns:

"Strong" if it contains at least : 1 UPPERCASE, 1 NUMBER, 1 SPECIAL CHARACTER, AND LENGTH >=8

"Moderate" if only 2 of those conditions are met

"Weak" otherwise

Ask the user for a password and display the result

```python
import string
def validate_pass(pas):
  conditions = 0
  if len(pas) >= 8:
    conditions += 1

  upper = False
  digit = False
  special = False

  for c in pas:
      if c.isupper():
          upper = True
      if c.isdigit():
          digit = True
      if c in string.punctuation:
          special = True

  if upper:
    conditions += 1
  if digit:
    conditions += 1
  if special:
    conditions += 1

  if conditions== 4:
    return "Strong"
  elif conditions >= 2 :
    return "Moderate"
  else:
    return "Weak"

validate_pass(input("Enter password: "))
```

```python
user_input = input("Enter your name: ")
while user_input.lower() != "exit":
  print(f"Welcome {user_input}")
  user_input = input("Enter your name: ")
```

```python
flag = True
while flag:
  name = input ("Enter your name: ")
  if name == "exit":
    flag = False
    print("Bye Have a good day! ")
  else:
    print(f"Welcome {name}")
```
